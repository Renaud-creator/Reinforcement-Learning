## <font color=darkblue> UCB1 Algorithm </font>

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from math import *
```

```python
bandit_probability = [0.2,0.5,0.75]
N_tries=10000

win_rate = np.zeros(N_tries)
reward_vector=np.zeros(N_tries)
total_reward = 0
num_optimal = 0
optimal_bandit = bandit_probability.index(max(bandit_probability))
```

```python
#returns 1 (success) with the probability of the slot machine picked up and 0 otherwise
# i is the parameter representing the slot machine which is played

def pull(i):
    global total_reward
    bandit_tries[i]+=1 
    est = random.random()
    if est < bandit_probability[i]:
        x = 1     #x is the reward of the slot machine which is either 0 or 1
        total_reward+=1
        return(x)
    else : 
        x = 0
        return(x)  
```

```python
#This function updates the estimated win rates of the slot machines, by calculating the samples mean

def update(i,x):
    bandit_est_prob[i] = (1/bandit_tries[i])*(bandit_est_prob[i]*(bandit_tries[i]-1)+x)
```

```python
def epsilon_greedy(param):
    global num_optimal, bandit_est_prob, bandit_tries, reward_vector
    bandit_tries=[1,1,1]
    bandit_est_prob=[pull(0), pull(1), pull(2)]
    reward_vector[0:3]=[pull(0), pull(1), pull(2)]
    for i in range (N_tries-3):
        L=[]
        for k in range(len(bandit_est_prob)):
            L.append(bandit_est_prob[k] + sqrt(param*log(i+3)/bandit_tries[k]))
        j = L.index(max(L))
        if j == optimal_bandit:
            num_optimal+=1
        reward = pull(j)
        reward_vector[i+3]=reward
        update(j,reward)
```

```python
param=2
epsilon_greedy(param)
```

```python
print("The total reward earned is "+str(total_reward))
print("The overall win rate is "+str(100*total_reward/N_tries)+"%")
print("The ratio of optimal choice meaning the number of times the maximum win rate slot machine was chosen divided by the number of times slot machines were pulled is "+str(100*num_optimal/N_tries)+"%")
print("The estimated mean for the first slot machine is "+str(bandit_est_prob[0])+" and the real one is "+str(bandit_probability[0]))
print("The estimated mean for the second slot machine is "+str(bandit_est_prob[1])+" and the real one is "+str(bandit_probability[1]))
print("The estimated mean for the third slot machine is "+str(bandit_est_prob[2])+" and the real one is "+str(bandit_probability[2]))
print("The number of times the first bandit is chosen is "+str(bandit_tries[0]))
print("The number of times the second bandit is chosen is "+str(bandit_tries[1]))
print("The number of times the third bandit is chosen is "+str(bandit_tries[2]))

win_rate=np.cumsum(reward_vector)/(np.arange(N_tries)+1)
optimal_win_rate=np.ones(N_tries)*max(bandit_probability)

plt.figure(figsize=(15, 9))
plt.plot(win_rate[:], 'r', lw=1.5, label='Win rate')
plt.plot(optimal_win_rate[:], 'g', lw=1.5, label='Optimal win rate')
plt.xlabel('Number of tries')
plt.ylabel('Win rate')
plt.legend()
plt.title("Evolution of the win rate through time with hyperparameter = "+str(param))
plt.show()
```

## <font color=darkblue> Conclusion and remarks </font>

We can see that the constant before the log is a hyperparameter designed to choose how much exploration we want. If we want little exploration, then we need to set a low value for this parameter, and if we want more exploration, we need to set this value higher. The number of times each bandit is selected highlights this statement. 

When the number of plays is high, then the second term in the argmax founction inside the greedy loop vanished because all bandit denominator $n_j$ will overtake the numerator $log(N)$ so that only the estimated mean remains. This means that exploration is over and that only exploitation takes place. 

Comparing to the optimistic initial value method, we can see that the suboptimal bandits' means are better estimated because not initialized with a high value.
